```python
import s3fs
```

```python
fs = s3fs.S3FileSystem()


my_files = fs.ls('ws-out/EastAfrica/Run03_10_2022/tileallea/2000/')
```

```python
my_files
```

```python
import re
def return_dailys(file_list, product_match, year):
    prune_list=[]
    expression = '.*' + product_match + str(year) + 'dek'
    for tif in file_list:
        match = re.match(expression, tif)
        if match:
            prune_list.append(tif)
    return prune_list
```

```python
year = 2000
product_match = 'etnow_'
a = return_dailys(my_files, product_match, year)
```

```python
a
```

```python
import os
for i in a:
    bn = os.path.basename(i)
    print(bn)
    from_path = f'ws-out/EastAfrica/Run03_10_2022/tileallea/2000/{bn}'
    to_path = f'ws-enduser/EastAfrica/2000/{bn}'
    print('cp', from_path, to_path)
    fs.cp(from_path, to_path)
    
```

```python
def replicate_year(year):
    my_files = fs.ls(f'ws-out/EastAfrica/Run03_10_2022/tileallea/{year}/')
    product_match = 'etnow_'
    a = return_dailys(my_files, product_match, year)
    for i in a:
        bn = os.path.basename(i)
        print(bn)
        from_path = f'ws-out/EastAfrica/Run03_10_2022/tileallea/{year}/{bn}'
        to_path = f's3{year}/{bn}'
        print('cp', from_path, to_path)
        fs.cp(from_path, to_path)
```

```python
replicate_year(2001)
```

```python
for year in range(2002, 2021):
    replicate_year(year)
```

```python
def replicate_yearV2(year, product_match):
    my_files = fs.ls(f'ws-out/EastAfrica/Run03_10_2022/tileallea/{year}/')
    #product_match = 'etnow_'
    a = return_dailys(my_files, product_match, year)
    for i in a:
        bn = os.path.basename(i)
        print(bn)
        from_path = f'ws-out/EastAfrica/Run03_10_2022/tileallea/{year}/{bn}'
        to_path = f'ws-enduser/EastAfrica/{year}/{bn}'
        print('cp', from_path, to_path)
        fs.cp(from_path, to_path)
```

```python
for product in ['netet_', 'etasw_', 'etc_']:
    for year in range(2000, 2021):
        replicate_yearV2(year, product)
```
